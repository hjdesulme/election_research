```python
import pandas as pd	

import datetime as dt
import matplotlib.dates as mdates 

import matplotlib.pyplot as plt

from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()
```

```python
#data source access
df=pd.read_csv('../Trend_CSV/Candidate-Democratic-Primary-New_York-Trend.csv')

#list setup
list_labels = list(df.columns.values.tolist())
list_x_values = list(df[list_labels[0]]) #x_axis Index [Dates]
list_1_values = list(df[list_labels[1]]) #y_axis Hillary Clinton: (New York) [Searches]
list_2_values = list(df[list_labels[2]]) #y_axis Bernie Sanders: (New York) [Searches]

#datetime converter
converted_dates = list(map(dt.datetime.strptime, list_x_values, len(list_x_values)*['%Y-%m-%d']))
x_axis = converted_dates

#chart setup
fig=plt.figure(figsize=(10, 5));

#plot(x data points, y data points, label)
plt.plot(x_axis,list_1_values,label=list_labels[1])
plt.plot(x_axis,list_2_values,label=list_labels[2])

#descriptive information
plt.title('Candidate-Democratic-Primary-New_York-Trend')
plt.xlabel('Day')
plt.ylabel('Interest over time')

#commands
plt.legend()
fig.autofmt_xdate()
plt.show()

#for testing:
#print(list_labels)
#print(list_x_values)
#print(len(list_x_values))
#df

plt.savefig('Candidate-Democratic-Primary-New_York-Line.png', format='png')    
plt.close()
```
