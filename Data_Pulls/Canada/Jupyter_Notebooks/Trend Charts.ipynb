```python
import pandas as pd	

import datetime as dt
import matplotlib.dates as mdates 

import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()
```

```python
#data source access
df=pd.read_csv('../Trend_CSV/Candidate-Federal-Canada-Trend.csv')

#list setup
list_labels = list(df.columns.values.tolist())
list_x_values = list(df[list_labels[0]]) #x_axis Index [Dates]
list_1_values = list(df[list_labels[1]]) #y_axis 
list_2_values = list(df[list_labels[2]]) #y_axis 
list_3_values = list(df[list_labels[3]]) #y_axis 

#datetime converter
converted_dates = list(map(dt.datetime.strptime, list_x_values, len(list_x_values)*['%Y-%m-%d']))
x_axis = converted_dates

#chart setup
fig=plt.figure(figsize=(10, 5));
ax=plt.axes()

#plot(x data points, y data points, label)
plt.plot(x_axis,list_1_values,label=list_labels[1])
plt.plot(x_axis,list_2_values,label=list_labels[2])
plt.plot(x_axis,list_3_values,label=list_labels[3])

#descriptive information
plt.title('Candidate-Federal-Canada-Trend')
plt.xlabel('Day')
plt.ylabel('Interest over time')

#commands
plt.legend()
fig.autofmt_xdate()
ax.xaxis.set_major_locator(ticker.MultipleLocator(10))
plt.show() #interferes with plt.savefig()

#plt.savefig('../Line_Chart_PNG/Candidate-Federal-Canada-Line.png', format='png') #saving as pngs atm
#plt.close()
```
